```python
import json
import pandas as pd

with open("2026-03-10_17-47-19.json", "r") as JSON:
    data_dict = json.load(JSON)

# Отбор нужных полей
df = pd.DataFrame(data_dict["hourly"])

# Приведение времени к datetime
df["time"] = pd.to_datetime(df["time"])

# Переименование столбцов
df.columns = ["Время, ГГГГ-ММ-ДД ЧЧ:ММ:СС", 
              "Температура, °C", 
              "Влажность, %",
              "Осадки, мм",
              "Скорость ветра, км/ч"
            ]

display(df)
print("Типы данных таблицы:\n", df.dtypes)
print("\nПервые 2 строки:\n", df.head(2))
print(f"\nСтрок: {df.shape[0]}, столбцов: {df.shape[1]}")

df.to_csv("2026-03-10_17-47-19.csv", sep=",", index=False, encoding="utf-8")
```
